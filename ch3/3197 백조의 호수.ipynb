```python
#3197 백조의 호수
from collections import deque

# r, c = 8, 17
# rake = [list('...XXXXXX..XX.XXX'),
#         list('....XXXXXXXXX.XXX'),
#         list('...XXXXXXXXXXXX..'),
#         list('..XXXXX.LXXXXXX..'),
#         list('.XXXXXX..XXXXXX..'),
#         list('XXXXXXX...XXXX...'),
#         list('..XXXXX...XXX....'),
#         list('....XXXXX.XXXL...')]

r, c = map(int,input().split())
rake = [list(input()) for _ in range(r)]

dy = [0,1,0,-1]
dx = [1,0,-1,0]
q = deque()
temp = deque()
flag = False
ice = deque()
melting = deque()
visit = [[0] * c for _ in range(r)]
visited = [[0] * c for _ in range(r)]
day = 0

def melt():
    while ice:
        y, x = ice.popleft()
        for i in range(4):
            ny = y + dy[i]
            nx = x + dx[i]
            if ny < 0 or nx < 0 or ny >= r or nx >= c or visit[ny][nx]: continue
            if rake[ny][nx] == 'X':
                visit[ny][nx] = 1
                melting.append((ny,nx))
                rake[ny][nx] = '.'
            

def bfs():
    global flag
    while q:
        y, x = q.popleft()
        for i in range(4):
            ny = y + dy[i]
            nx = x + dx[i]
            if ny < 0 or nx < 0 or ny >= r or nx >= c or visited[ny][nx]: continue
            if rake[ny][nx] == '.': q.append((ny,nx))
            elif rake[ny][nx] == 'X':
                temp.append((ny,nx))
            elif rake[ny][nx] == 'L':
                flag = True
                return
            visited[ny][nx] = 1


for i in range(r):
    for j in range(c):
        if rake[i][j] == 'L':
            y = i
            x = j
        if rake[i][j] == '.':
            visited[i][j] = 1
            ice.append((i,j))


q.append((y,x))
while flag == False:
    visited[y][x] = 1
    bfs()
    if flag: break
    melt()
    ice = melting
    q = temp
    melting = deque()
    temp = deque()
    day += 1
    

print(day)


```
