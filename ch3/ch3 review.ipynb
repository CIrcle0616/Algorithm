```python
from collections import deque

n = int(input())
scv = list(map(int, input().split()))
while len(scv) < 3:
    scv.append(0)
attack_variation = [[9, 3, 1],
                    [9, 1, 3],
                    [3, 9, 1],
                    [3, 1, 9],
                    [1, 3, 9],
                    [1, 9, 3]]
hp = [[[0] * 61 for _ in range(61)] for _ in range(61)]
a, b, c = scv[0], scv[1], scv[2]

def attack(a, b, c):
    q = deque()
    q.append((a, b, c))
    hp[a][b][c] = 1
    while q:
        if hp[0][0][0]: break
        a, b, c = q.popleft()
        for i in range(6):
            a_attacked, b_attacked, c_attacked = a - attack_variation[i][0], b - attack_variation[i][1], c - attack_variation[i][2]
            a_attacked, b_attacked, c_attacked = max(0, a_attacked), max(0, b_attacked), max(0, c_attacked)
            if hp[a_attacked][b_attacked][c_attacked]: continue
            hp[a_attacked][b_attacked][c_attacked] = hp[a][b][c] + 1
            q.append((a_attacked, b_attacked, c_attacked))
attack(a, b, c)

print(hp[0][0][0] - 1)
            


```

```python
n = int(input())
nums = list(input())
max_value = -10e9

def calc(a, b, op):
    if op == "+":
        return int(a) + int(b)
    if op == "-":
        return int(a) - int(b)
    if op == "*":
        return int(a) * int(b)

def dfs(idx, ans):
    global max_value
    if idx == n-1: 
        max_value = max(max_value, int(ans))
        return
    dfs(idx + 2, calc(ans, nums[idx + 2], nums[idx + 1]))
    if idx + 4 <= n-1:
        temp = calc(nums[idx + 2], nums[idx + 4], nums[idx + 3])
        dfs(idx + 4, calc(ans, temp, nums[idx + 1]))

dfs(0, nums[0])
print(max_value)
```

```python
#주난의난
from collections import deque
dy, dx = [1, -1, 0, 0], [0, 0, 1, -1]

n, m = map(int, input().split())
y, x, _y, _x = map(int, input().split())
a = [list(input()) for _ in range(n)]

# n, m = 5, 7
visited = [[0] * m for _ in range(n)]
# y, x, _y, _x = 3, 4, 1, 2
# a = [list('1#10111'),
# list('1101001'),
# list('001*111'),
# list('1101111'),
# list('0011001')]
cnt = 1
def bfs(y, x):
    global cnt
    q = deque()
    _q = deque()
    q.append((y, x))
    visited[y][x] = 1
    while q:
        y, x = q.popleft()
        for i in range(4):
            ny, nx = y + dy[i], x + dx[i]
            if 0 <= ny < n and 0 <= nx < m and visited[ny][nx] == 0:
                visited[ny][nx] = 1
                if a[ny][nx] == "1":
                    _q.append((ny, nx))
                elif a[ny][nx] == "0":
                    q.append((ny, nx))
                elif a[ny][nx] == "#":
                    return
        if len(q) == 0:
            q = _q
            cnt += 1
            _q = deque()
bfs(y-1, x-1)
print(cnt)
```

```python
#백조의호수
from collections import deque

dy, dx = [1, -1, 0, 0], [0, 0, 1, -1]

r, c = map(int, input().split())
a = [list(input()) for _ in range(r)]
ice = deque()
visited = [[0] * c for _ in range(r)]
visited_L = [[0] * c for _ in range(r)]
tempSwan = deque()
waterQ = deque()
swanQ = deque()

def bfs(queue):
    while queue:
        y, x = queue.popleft()
        for i in range(4):
            ny, nx = y + dy[i], x + dx[i]
            if ny < 0 or ny >= r or nx < 0 or nx >= c or visited_L[ny][nx]: continue
            visited_L[ny][nx] = 1
            if a[ny][nx] == '.': queue.append((ny, nx))
            elif a[ny][nx] == 'X':
                tempSwan.append((ny,nx))
            elif a[ny][nx] == 'L':
                return True

def melt(waterQ):
    while waterQ:
        y, x = waterQ.popleft()
        for i in range(4):
            ny, nx = y + dy[i], x + dx[i]
            if ny < 0 or ny >= r or nx < 0 or nx >= c or visited[ny][nx]: continue
            visited[ny][nx] = 1
            if a[ny][nx] == 'X':
                ice.append((ny, nx))
                continue
            waterQ.append((ny, nx))
    for y, x in ice:
        a[y][x] = '.'

day = 0
for i in range(r):
    for j in range(c):
        if a[i][j] == 'L':
            y, x = i, j
        if a[i][j] == '.' or a[i][j] == 'L':
            waterQ.append((i, j))
swanQ.append((y, x))
visited_L[y][x] = 1
while True:
    if bfs(swanQ): break
    melt(waterQ)
    waterQ = ice
    swanQ = tempSwan
    ice = deque()
    tempSwan = deque()
    day += 1

print(day)
    

    
```
