```python
#13913 숨바꼭질4
from collections import deque

n, k = map(int,input().split())
M = 200004
visited = [0] * M
prev =  [0] * M

if n == k:
    print(0)
    print(n)
else:
    visited[n] = 1
    q = deque()
    q.append(n)
    while q:
        now = q.popleft()
        if now == k: 
            ret = visited[k]
            break
        for next in [now+1, now-1, now*2]:
            if next < 0 or next >= M or visited[next]: continue
            visited[next] = visited[now] + 1
            prev[next] = now
            q.append(next)
            

    stack = []
    i = k
    while i != n:
        stack.append(i)
        i = prev[i]
    stack.append(n)
    print(visited[k]-1)
    while stack:
        print(stack.pop(), end=' ')

```
