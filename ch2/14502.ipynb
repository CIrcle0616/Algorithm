```python
#14502 연구소
from collections import deque
from itertools import combinations
import copy
import time

dy = [0,1,0,-1]     
dx = [1,0,-1,0]

def bfs(y,x):
    q.append((y,x))    
    while q:
        y,x = q.popleft()
        graph_copy[y][x] = 2
        for i in range(4):
            ny = y + dy[i]
            nx = x + dx[i]
            if ny >= N or nx >= M or ny < 0 or nx < 0:
                continue
            if graph_copy[ny][nx] != 0:
                continue
            q.append((ny,nx))
            

N, M = map(int,input().split())
graph = [list(map(int,input().split())) for _ in range(N)]
big = 0
q = deque()
coordinates = []     #좌표지정
chc = 0
for y in range(N):
    for x in range(M):
        if graph[y][x] == 0:
            coordinates.append([y,x])
for c in combinations(coordinates,3):
    start = time.time()
    graph_copy = copy.deepcopy(graph)
    for y,x in c:
        graph_copy[y][x] = 1
    virus = deque()
    for i in range(N):
        for j in range(M):
            if graph_copy[i][j] == 2:
                virus.append((i,j))
    while virus:
        y,x = virus.popleft()
        bfs(y,x)
    cnt = 0
    for i in range(N):
        cnt += graph_copy[i].count(0)
        if cnt > big:
            big = cnt
    end = time.time()
    chc += 1
    print(chc, end - start)
print(big)
```
